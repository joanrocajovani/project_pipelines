### Which directors have been more succesful?

```python
director_count = df["Director"].value_counts()[:10].index.to_list()
director = pd.Categorical(df['Director'], director_count)
sns.set(rc={'figure.figsize':(12,8)})
sns.histplot(data=df, x=director, multiple="stack").set(title= "Top 1000 movies by director")
plt.xticks(rotation=60)
plt.plot()
```

```python
director_count = df_oscars["Director"].value_counts()[:10].index.to_list()
director = pd.Categorical(df_oscars['Director'], director_count)
sns.set(rc={'figure.figsize':(12,8)})
sns.histplot(data=df_oscars, x=director, multiple="stack").set(title= "Best Picture movies by director")
plt.xticks(rotation=60)
plt.plot()
```

```python
director_series = df.groupby("Director")["Gross_(M)"].sum()
```

```python
director_series.sort_values(ascending=False)[:10].plot(kind='bar' ,title='Gross Made by Director')
```
